```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
df=pd.read_csv('MSFT.csv')
df=df.reset_index()['Close']
```

```python
df = np.array(df).reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
df = scaler.fit_transform(df)
```

```python
tr_sz=int(len(df)*0.6)
val_sz=int(len(df)*0.15)
te_sz=len(df)-tr_sz-val_sz
train_data= df[:tr_sz,:]
val_data=df[tr_sz:tr_sz+val_sz,:]
test_data=df[tr_sz+val_sz:,:]
```

```python
def hlp(data,nof=1):
    x=[]
    y=[]
    for i in range(len(data)-nof-1):
        a = data[i:(i+nof), 0] 
        x.append(a)
        y.append(data[i+nof+1,0])
    return np.array(x),np.array(y);
```

```python
xtr,ytr=hlp(train_data,70)
xval,yval=hlp(val_data,70)
xts,yts=hlp(test_data,70)
```

```python
xtr=xtr.reshape(xtr.shape[0],xtr.shape[1],1)
xval=xval.reshape(xval.shape[0],xval.shape[1],1)
xts=xts.reshape(xts.shape[0],xts.shape[1],1)
```

```python
model=Sequential(
[LSTM(40,return_sequences=True,input_shape=(70,1)),
 LSTM(40,return_sequences=True),
 LSTM(40,return_sequences=True),
 LSTM(40),
 Dense(1)
]
)
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(xtr,ytr,validation_data=(xval,yval),epochs=50,batch_size=32,verbose=1)
```

```python
ts_pred=model.predict(xts)
```

```python
ts_pred=scaler.inverse_transform(ts_pred)
```
